Let's check further assumptions

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler()
```

```python
X = df.iloc[:,1:-3]
y = df['House price of unit area']
```

```python
Xc = sc.fit_transform(X)
```

```python
X = pd.DataFrame(Xc, columns = X.columns)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.2, random_state=500)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error, r2_score,mean_absolute_percentage_error
```

```python
mean_absolute_percentage_error(y_test, y_pred)
```

```python
mean_squared_error(y_test, y_pred)
```

```python
r2_score(y_test, y_pred)
```

```python
residuals = y_test.values-y_pred
mean_residuals = np.mean(residuals)
print(mean_residuals)
```

here mean of the residuels should be zero or atleast near to zero. But we are having 2.55. So it's failing mean of residual assumptions too.

we will use randomforrest regressor 

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
model = RandomForestRegressor()
```

```python
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
mean_absolute_percentage_error(y_test, y_pred)
```

```python
mean_squared_error(y_test, y_pred)
```

```python
r2_score(y_test, y_pred)
```

```python
residuals = y_test.values-y_pred
mean_residuals = np.mean(residuals)
print(mean_residuals)
```

Here when you compare both model, Random forrest regressor is outperforming Linear regression.
